```python
results = []
for li in lis:
    news_title = li.find('a').text.strip()
    news_link = li.find('a').attrs['href']
    print(news_title, news_link)
    results.append((news_title, news_link))
results
```

```python
url = "https://www.rottentomatoes.com/"
response = requests.get(url)
```

```python
response
```

```python
response.text
```

```python
html = response.text
```

```python
soup = BeautifulSoup(html, 'html.parser') 
```

```python
table = soup.find('table',attrs={'id':'Top-Box-Office'})
```

```python
table
```

```python
trs = table.find_all('tr')
```

```python
trs[0]
```

```python
score = trs[0].find('span',attrs={'class':'tMeterScore'}).text[:2]

score
```

```python
title = trs[0].find('td',attrs={'class':'middle_col'}).text.replace('\n','')
title
```

```python
link = url + trs[0].find('a').attrs['href']
link
```

```python
revenue= trs[0].find('td',attrs={'class':'right_col right'}).text.replace('\n','')
revenue
```

```python
results = []
for tr in trs:
    score = tr.find('span',attrs={'class':'tMeterScore'}).text[:2]
    title = tr.find('td',attrs={'class':'middle_col'}).text.replace('\n','')
    revenue= tr.find('td',attrs={'class':'right_col right'}).text.replace('\n','')
    link = url + tr.find('a').attrs['href']
    print(score,title,revenue,link)
    results.append((score,title,revenue,link))
```
